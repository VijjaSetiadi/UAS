```python
data1 = pd.read_csv("stock.csv")
data1.head()
```

a. Distribusi data pada setiap atribut Purchase Price, Current Price dan Gain/Loss dengan grafik histogram

```python
sns.histplot(y=data1["Current Prices"])
```

```python
sns.histplot(y=data1["Gain/Loss"])
```

b. Atribut Stock Name dan Symbol yang memiliki Dividen / Share 5 tertinggi

```python
dividendshare = (data1.groupby("Stock Name")["Dividend/share"]
.sum()
.reset_index()
.sort_values('Dividend/share',ascending=False)
.head(5))
print(dividendshare)

import seaborn as sns
sns.catplot(x=dividendshare["Dividend/share"],y=dividendshare["Stock Name"],data=dividendshare,height=5, aspect=11.7/8.27)
```

c. Atribut Stock Name dan Symbol yang memiliki Market Value 5 tertinggi

```python
marketv = (data1.groupby("Stock Name")["Market Value"]
.sum()
.reset_index()
.sort_values('Market Value',ascending=False)
.head(5))
print(marketv)


import seaborn as sns
sns.catplot(x=marketv["Stock Name"],y=marketv["Market Value"],data=marketv,height=5, aspect=20/8.27)
```

d. Atribut Stock Name dan Symbol yang memiliki Annual Yield 7 tertinggi

```python
import matplotlib.pyplot as plt
annualy = (data1.groupby("Stock Name")["Annual Yield"]
.sum()
.reset_index()
.sort_values('Annual Yield',ascending=False)
.head(7))
print(annualy)

sns.catplot(x=annualy["Stock Name"],y=annualy["Annual Yield"],data=annualy,height=5, aspect=20/8.27)
```
